# Etapa 1: Importação das bibliotecas

```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random 
import zipfile
tf.__version__
```

# Etapa 2: Importação do modelo com os pesos pré-treinados

```python
model = tf.keras.applications.ResNet50(weights = 'imagenet')
```

# Etapa 3: Testes com o modelo pré-treinado

```python
sample_elefante = tf.keras.preprocessing.image.load_img(r'../content/Transfer Learning Data/African_Elephant.jpg', target_size=(224,224))
```

# Etapa 3.1: Elefante

```python
plt.imshow(sample_elefante);
```

```python
sample_elefante = tf.keras.preprocessing.image.img_to_array(sample_elefante)
sample_elefante = np.expand_dims(sample_elefante, axis = 0)
sample_elefante = tf.keras.applications.resnet50.preprocess_input(sample_elefante)
predictions_elefante = model.predict(sample_elefante)
print('predictions: ', tf.keras.applications.resnet50.decode_predictions(predictions_elefante, top = 5))
```

# Etapa 3.2: Neninha

```python
sample_neninha = tf.keras.preprocessing.image.load_img(
    r'../content/Transfer Learning Data/neninha2.jpeg', target_size=(224,224)
    )
plt.imshow(sample_neninha);
```
